# First order ODE:
Here in this notebook we are solving an one dimensional ordinary differential equation. The equation is as following:

$y'= x - 2y$  where $0 \le x \le 1$ and the boundry condition is  $y(0) = 3$. Using the following code we solve it numerically and compare it to it's theoretical solution which is easy to obtain.

```python
import numpy as np 
# Form scipy we will import ODE solver here
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Defining th function
# The order of y and x matters !!!!
def yprime(y,x):
    return x-2.*y

# Theortical solution 
def function(x):
    return 0.5*x - 0.25 + 3.25 * np.exp(-2.*x)

# The initial condition 
x_range = np.linspace(0.,1.,1000)
y0=3.


#Solving the ODE here 
sol = odeint(yprime,y0,x_range)

# Here we plot the numerical solution
plt.plot(x_range,sol,color = 'red', label = 'Numerical solution')

# Ploting the theoretical solution, it has been shifted up by 0.1
# just for the purpose of the demonestration
plt.plot(x_range,function(x_range) + 0.1,color = 'blue', label = 'Theoretical solution')

# Defining the legend and axis labels and font sizes 
plt.legend(loc=1,fontsize = 14)
plt.xlabel('x',fontsize = 20)
plt.ylabel('y',fontsize = 20)
plt.tick_params(axis='both', labelsize=12)

# Showing the plot 
plt.show()

```
